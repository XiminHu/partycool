```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image = cv2.imread('sem_1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
print(image.shape)
height = image.shape[0]
width = image.shape[1]
```

```python
region_of_interest_array = [(width/2, height),(width,height/2),(width, height)]
```

```python
def region_of_interest(img, vertices):
    """
    Mask uninterested region
    """
    mask = np.zeros_like(img)
    channel_count = img.shape[2]
    match_mask_color = (255,)*channel_count
    cv2.fillPoly(mask, vertices, match_mask_color)
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
```

```python
cropped_image = region_of_interest(image, np.array([region_of_interest_array], np.int32))
```

```python
plt.figure(figsize=(15,15))
plt.imshow(cropped_image)
plt.show()
#plt.axis('off')
```
